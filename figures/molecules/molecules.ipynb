# making molecule pictures

```python
from xyzrender import load, render, render_gif

mol = load("CC(=O)OC", smiles=True)
render(mol)                          # displays inline in Jupyter
render(mol, output="methylacetate.png")   # save without requiring cairo


mol = load("CC(=O)OCC", smiles=True)
render(mol)                          # displays inline in Jupyter
render(mol, output="ethylacetate.png", orient=True)   # save without requiring cairo


mol = load("CCO", smiles=True)
render(mol)                          # displays inline in Jupyter
render(mol, output="ethanol.png", orient=True)   # save without requiring cairo
```
